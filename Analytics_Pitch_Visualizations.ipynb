```python
import matplotlib.pyplot as plt
import numpy as np

# Years for the X-axis
years = np.array([2019, 2020, 2021, 2022, 2023])

# Hypothetical percentage of revenue lost to food spoilage for the Y-axis
revenue_loss_percentage = np.array([20, 21, 19, 22, 27])

# Creating the bar chart with the requested adjustments
plt.figure(figsize=(10, 6))
plt.bar(years, revenue_loss_percentage, color='grey')  # Grey bars

# Highlighting the 2023 data with blue, but not adding a legend this time
plt.bar(years[-1], revenue_loss_percentage[-1], color='blue')

# Adding titles and labels
plt.title('Annual Impact of Food Spoilage on Revenue', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Revenue Loss Due to Spoilage (%)', fontsize=12)
plt.xticks(years)
plt.ylim(0, 40)  # Scaling the Y-axis to 40

# Adding the percentage values above each bar
for i, loss in enumerate(revenue_loss_percentage):
    plt.text(years[i], loss + 0.5, f"{loss}%", ha='center', va='bottom')

# Displaying the plot without a legend
plt.tight_layout()
plt.show()
```
